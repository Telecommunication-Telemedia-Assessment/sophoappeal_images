```python
%matplotlib inline

from IPython.display import display as d
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(rc={'figure.figsize':(12,12)})
sns.set_style("white")
import glob
import json
import os
import sys
import shutil
import multiprocessing
pool = multiprocessing.Pool(multiprocessing.cpu_count())

def save_fig(ax, title, pdfname):
    ax.set_title(title)
    ax.get_figure().savefig(pdfname, bbox_inches="tight")
```

```python
dna = pd.read_json("nima_appeal.json")
d(dna)

sns.set(rc={'figure.figsize':(8,5)})
sns.set_style("white")
ax = sns.distplot(dna["mean_score_prediction"])

save_fig(ax, "nima appeal prediction", "distplot_nima_appeal.pdf")
```
